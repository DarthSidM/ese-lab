```python
import numpy as np
from scipy.stats import friedmanchisquare, wilcoxon
import itertools
```

```python
# Rows = datasets (5)
# Columns = models [LR, DT, RF, XGB]

accuracies = np.array([
    [0.80, 0.75, 0.85, 0.88],  # Dataset 1
    [0.60, 0.65, 0.70, 0.72],  # Dataset 2
    [0.90, 0.85, 0.92, 0.95],  # Dataset 3
    [0.50, 0.55, 0.60, 0.65],  # Dataset 4
    [0.70, 0.68, 0.75, 0.78]   # Dataset 5
])
```

```python
# Split columns (models)
lr = accuracies[:, 0]
dt = accuracies[:, 1]
rf = accuracies[:, 2]
xgb = accuracies[:, 3]

stat, p_value = friedmanchisquare(lr, dt, rf, xgb)

print("Friedman Test Statistic:", stat)
print("Friedman p-value:", p_value)

if p_value < 0.05:
    print("Significant difference exists between models")
else:
    print("No significant difference between models")
```

```python
model_names = ["LR", "DT", "RF", "XGB"]

for (i, j) in itertools.combinations(range(4), 2):
    stat, p = wilcoxon(accuracies[:, i], accuracies[:, j])
    print(f"{model_names[i]} vs {model_names[j]}: p-value = {p}")
```
